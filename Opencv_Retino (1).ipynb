```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage.color import rgb2gray

#read image
img1 = plt.imread('C:/Users/PB/Desktop/imagep/retino.jpg')

img = cv2.imread('C:/Users/PB/Desktop/imagep/retino.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

#applying ClAHE
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
cl1 = clahe.apply(gray)
#dilation
kernel = np.ones((5,5),np.uint8)
closing = cv2.morphologyEx(cl1, cv2.MORPH_CLOSE, kernel)

#hsv = cv2.cvtColor(closing, cv2.COLOR_RGB2HSV)
#thresholding
ret,th1 = cv2.threshold(closing,127,255,cv2.THRESH_BINARY)
median = cv2.medianBlur(th1,5)

plt.imshow(img1)
plt.show()
plt.imshow(median,'gray')
plt.show()
```
